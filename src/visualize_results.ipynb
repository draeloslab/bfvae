```python
import numpy as np
import matplotlib.pyplot as plt

def plot_relevance(pv):
    ri = np.abs(1-pv)
    x_labels = list(range(1, 11))

    plt.figure(figsize=(10, 1))
    plt.bar(x_labels, ri)
    plt.title('Relevance indicator (r)')
    plt.xticks(x_labels)
    plt.yticks([])

    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(True)
    ax.spines['bottom'].set_visible(True)

    plt.show()

pv = np.array([1.06, 0.99, 0.99, 1.08, 1.05, 0.99, 1., 1.15, 1.05, 0.97])
plot_relevance(pv)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

def plot_traversal(model_id, experiment_id, checkpoint_id, traversal_id):
    image_ids = np.arange(0, 70, 10)
    image_paths = [f'{model_id}/outputs/{experiment_id}/{checkpoint_id}/{traversal_id}_{ix:03}.jpg' for ix in image_ids]
    images = [Image.open(img_path) for img_path in image_paths]
    width, height = images[0].size
    total_height = height * len(images)
    combined_image = Image.new('RGB', (width, total_height))
    y_offset = 0
    for img in images:
        combined_image.paste(img, (0, y_offset))
        y_offset += height
    combined_image_np = np.array(combined_image)
    plt.imshow(combined_image_np)
    plt.axis('off')
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
    plt.show()

model_id = 'bfvae0'
experiment_id = 'oval_dsprites_eta_1.0_gamma_6.4_zDim_10_lrVAE_0.0001_lrD_0.0001_rseed_0_run_0_trvsl'
checkpoint_id = '300000'
traversal_id = 'fixed1'
plot_traversal(model_id, experiment_id, checkpoint_id, traversal_id)
```
